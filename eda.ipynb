### A.Univariant analysis ###

```python
for f in num_features:
    fig, axes = plt.subplots(1, 2, figsize=(12, 4)) 
    
    sns.histplot(data=df, x=f, ax=axes[0], kde=True)
    axes[0].set_title(f'Distribution of {f}')
    
    sns.boxplot(data=df, x=f, ax=axes[1])
    axes[1].set_title(f'Outliers in {f}')
    
    plt.tight_layout()
    plt.show()
```

### B.Bivariant analysis ###

```python
for f in num_features:
    for j in num_features:
        if f != j:
            cor = df[f].corr(df[j])
            if cor > 0.50:
                fig, ax = plt.subplots(figsize=(6, 4))
                plt.hexbin(data=df, x=f, y=j)
                ax.set_title(f'{f} vs {j} (r = {cor:.2f})')
                plt.show()
```

## 7.PCA components ##

```python
from sklearn.cluster import KMeans
from sklearn.preprocessing import MinMaxScaler

scaler = MinMaxScaler()

x_scaled = scaler.fit_transform(x)
from sklearn.decomposition import PCA

# 1. Initialize PCA without a limit first
pca = PCA()
pca.fit(x_scaled)

# 2. Calculate Cumulative Variance
cumulative_variance = np.cumsum(pca.explained_variance_ratio_)

# 3. Find the number of components needed for 90% variance
n_components = np.argmax(cumulative_variance >= 0.90) + 1
print(f"To explain 90% of the variance, use {n_components} components.")
```

```python
from sklearn.decomposition import PCA

x_scaled = scaler.fit_transform(x)

pca = PCA(n_components=6)

x_pca = pca.fit_transform(x_scaled)


print(pca.explained_variance_ratio_)
```

## 8.Elbow method ##

```python

inertia = []

for k in range(1,11):
    kmeans = KMeans(n_clusters=k,random_state=42)
    kmeans.fit(x_scaled)
    inertia.append(kmeans.inertia_)

plt.plot(range(1,11),inertia,marker='o')
plt.xlabel("number of cluster k")
plt.ylabel("inertia")
plt.title("elbow method")
plt.show()
```

## 9.KMeans Cluster ##

```python
from sklearn.cluster import KMeans

x_scaled = scaler.fit_transform(x)

kmean = KMeans(n_clusters=4,random_state=42,n_init="auto")

clusters = kmean.fit_predict(x_scaled)

print(kmeans.cluster_centers_)


```

```python
clusters
```

## 10. PCA Visualization ##

```python
import matplotlib.pyplot as plt
import seaborn as sns


plt.figure(figsize=(10, 7))
sns.scatterplot(
    x=x_pca[:, 0], 
    y=x_pca[:, 1], 
    hue=clusters, 
    palette='viridis', 
    s=60, 
    alpha=0.8
)

plt.title('K-Means Clustering on PCA-Reduced Data')
plt.xlabel('Principal Component 1')
plt.ylabel('Principal Component 2')
plt.legend()
plt.show()
```

## 11.Silhouette Score ##

```python
from sklearn.metrics import silhouette_score

label = clusters 

score = silhouette_score(x_scaled,label)

print(score)
```

## 12.Davies-Bouldin Index Score ##

```python
from sklearn.metrics import davies_bouldin_score

label = clusters

db_index=davies_bouldin_score(x_scaled,label)

print(f"Davies-Bouldin Index: {db_index}")
```

## 13.Add clusters to df and visualization ##

```python
df['clusters'] = kmean.labels_
```

```python
df.groupby('clusters').mean(numeric_only=True)
```

```python
import matplotlib.pyplot as plt


features_to_show = ['danceability', 'energy', 'acousticness', 'instrumentalness', 'valence','speechiness','liveness']

cluster_means = df.groupby('clusters').mean(numeric_only=True)

cluster_means[features_to_show].plot(kind='bar', figsize=(10, 6))

plt.title('Average Feature Values per Cluster')
plt.xlabel('Cluster')
plt.ylabel('Mean Value')
plt.xticks(rotation=0)
plt.legend(title='Features', bbox_to_anchor=(1.05, 1), loc='upper left')
plt.show()
```
